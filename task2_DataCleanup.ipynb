#**Task 2: Data Cleanup**

The provided `sampledata.csv` dataset is cleaned below using Python by following the instructions that are given in R.


```python
#importing neccessary libraries that are required to clean the given dataset

import pandas as pd
```

```python
#reading the CSV file into a dataframe

df = pd.read_csv('sampledata.csv')
```

```python
#displaying the top 5 and the bottom 5 rows of the dataframe

df.head()
```

```python
df.tail()
```

```python
#storing columns that need to be shortened and remained in respective variables:

cols_to_short = df.columns[df.columns.get_loc("new_sp_m014"):]
cols_to_rem = df.columns[:5]

print(cols_to_short)
print(cols_to_rem)
```

```python
#using the melt function from Pandas to shorten/ pivot the dataframe:

df_modified = pd.melt(df, id_vars= cols_to_rem, value_vars = cols_to_short, var_name='key', value_name='cases')

#dropping rows with missing vales in the 'cases' column:

df_modified = df_modified.dropna(subset=['cases'])

print(df_modified)
```

```python
#Printing the structure of the 'key' column

print(df_modified['key'].value_counts())
```

```python
#Since the variables are inconsistent in 'key', we have to make changes. The 'newrel' should be represented as 'new_rel'

for i in range(len(df_modified['key'])):
    if df_modified['key'].iloc[i].startswith('newrel'):
        df_modified['key'].iloc[i] = 'new_rel' + df_modified['key'].iloc[i][6:]

print(df_modified['key'].value_counts())
```

```python
#transforming the type, age, sex, and new values in a clean and readble columns

df_modified[['new', 'type', 'sexage']] = df_modified['key'].str.split('_', expand = True)

df_modified.head()
```

```python
#Since we don't need 'key' and 'new' now, lets drop it:

df_modified = df_modified.drop(columns=['key', 'new'])
```

```python
#dropping iso2 and iso3 too as they are the short form of the country:
df_modified = df_modified.drop(columns=['iso2', 'iso3'])
```

```python
#seperating sex and age from sexage:

df_modified['sex'] = df_modified['sexage'].str[0]
df_modified['age'] = df_modified['sexage'].str[1:]

df_modified = df_modified.drop(columns='sexage')
df_modified.tail()
```
